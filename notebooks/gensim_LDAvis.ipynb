```python
import pyLDAvis.gensim_models
import gensim
import pyLDAvis
from topic_modelling.data_loader import DataLoader
import gensim.corpora as corpora


data_path = "../data/abcnews_lem_stem.pickle"
dl = DataLoader()
n_rows = 100000
corpus = dl.load(data_path, n_rows=n_rows)

num_topics = 9
```

```python
corpus_words = corpus.values
corpus_words = [[word for word in doc if len(word) != 1] for doc in corpus_words]
corpus_words[:5]
```

```python
id2word = corpora.Dictionary(corpus.values)
id2word[4]
```

```python
corpus = [id2word.doc2bow(text) for text in corpus_words]
corpus[:5]
```

```python
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=num_topics, 
                                           random_state=100,
                                           chunksize=100,
                                           passes=10)
```

```python
from gensim.models import CoherenceModel
coherence_model_lda = CoherenceModel(model=lda_model, texts=corpus_words, dictionary=id2word, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
coherence_lda
```

```python
for t in range(num_topics):
    print(lda_model.show_topic(t, 5))
```

```python
# Visualize the topics
pyLDAvis.enable_notebook()
LDAvis_prepared = pyLDAvis.gensim_models.prepare(lda_model, corpus, id2word)
LDAvis_prepared
```
